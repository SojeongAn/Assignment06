# Definition

## Import packages

```python
import numpy as np
import statistics 
import matplotlib.pyplot as plt
```

## Declare the variable

```python
num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b  #noise data
y2      = a * x + b #clean line
```

## Visualization of noise/clean data: $Y_2 = 2x + b$

```python
plt.plot(x, y1, 'b.', label='Noisy data')
plt.plot(x, y2, 'r', label='Fitted line data')
plt.legend()
plt.show()
```

#  Least Squares Approximation

## Function of computing average: $f(x)= \frac{1}{N}1^T y$

```python
def computeAvg(vec):
    return np.mean(vec)
```

## Function of computing standard deviation:  $f(x)= \sqrt{\sum{}^{}\frac{(x - x')^2}{N}}$

```python
def computeStd(vec):
    return statistics.stdev(vec)
```

## Solution of Least Squares Approximation: $f(x)= avg(y)+ a\frac{std(y)}{std(x)}(x-avg(x))$

```python
def straightLineFit(vec1, vec2):
    return computeAvg(vec2) + (a * (computeStd(vec2)/computeStd(vec2)) * (x-computeAvg(vec1)))
```

## Visualization of the result

```python
y = straightLineFit(x,y1) #Least Squares Approxiamtion
plt.plot(x, y1, 'b.', label='Noisy data')
plt.plot(x, y, 'r', label='straight Line Fit')
plt.legend()
plt.show()
```
