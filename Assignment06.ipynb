# Definition

## Import packages

```python
import numpy as np
import statistics 
import matplotlib.pyplot as plt
from numpy.linalg import *
```

## Declare the variable

```python
num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b  #noise data
y2      = a * x + b #clean line
```

## Visualization of noise/clean data: $Y_2 = 2x + b$

```python
plt.plot(x, y1, 'b.', label='Noisy data')
plt.plot(x, y2, 'r', label='Fitted line data')
plt.legend()
plt.show()
```

# Least Squares Approximation

## Function of Least Squares Approximation: 

$f(x)= \sum_{i=1}^{m}(y_i - {x_i}^T W)^2 $

$ W = (X^T X)^{-1} X^T y $ 

$W = $ weight

```python
def straightLine(vecX,vecY):
    vecX = np.mat(vecX)
    vecY = np.mat(vecY).T
    xTx = vecX.T * vecX
    if(np.linalg.det(xTx) == 0.0): # if Singular function, return 
        print("This matrix is singular")
        return
    weight = xTx.I * (vecX.T * vecY) # weight
    return weight
```

## Computing regression equation

```python
interX = []
for i in range(0, len(x)):
    interX.append(np.array((np.ones_like(1), x[i])))  # Set X0 = 1.0 for computing

interX = np.mat(interX)
weight = straightLine(interX, y1)
a = float(weight[1]); b = float(weight[0]) # set the a, b 
y = a * x + b
plt.plot(x, y1, 'b.', label='Noisy data')
plt.plot(x, y, 'r', label='straight Line Fit')
```
